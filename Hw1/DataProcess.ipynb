```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# plt.rcParams['font.sans-serif'] = 'Arial Unicode MS'

# plt.rcParams['axes.unicode_minus'] = False
df = pd.read_csv("110_studentsc.csv") # read in data 

# deal with missing or null values
df = df.replace('-', 0, regex=True)
df = df.replace(',', '', regex=True)
```

```python
# Casting the type from object to string
df['總計'] = df['總計'].astype(str)
# Remove commas and convert the '總計' column to integer
df['總計'] = df['總計'].str.replace(',', '').astype(int)
total_students_by_major = df.groupby('細學類名稱')['總計'].sum()
print(total_students_by_major)

total_students_by_major.plot(kind='bar', figsize=(12, 6))
plt.title('Total Students by Major')
plt.xlabel('Major')
plt.ylabel('Total Students')
plt.xticks(rotation=90)
plt.show()
```

### 各等級別科系數量

```python
df1 = df.groupby('等級別')['科系名稱'].nunique().reset_index(name='科系數量')

print(df1)

# 繪製長條圖
plt.figure(figsize=(10, 6))
plt.bar(df1['等級別'], df1['科系數量'])
plt.title('各等級別科系數量')  # 添加標題
plt.xlabel('等級別')  # 添加 x 軸標籤
plt.ylabel('科系數量')  # 添加 y 軸標籤

# 顯示圖表
plt.tight_layout()
plt.show()
```

### 各等級別學制類別比例

```python
grouped_data = df.groupby(['等級別', '日間∕進修別']).size().reset_index(name='計數')

grouped_data
# 創建一個新的 DataFrame 以存儲比例
df_ratio = pd.DataFrame(columns=['等級別', '日間∕進修別', '比例'])

# Calculate the total counts for each level
total_counts = grouped_data.groupby('等級別')['計數'].transform('sum')

df_ratio['等級別'] = grouped_data['等級別']
df_ratio['日間∕進修別'] = grouped_data['日間∕進修別']

# Calculate the ratios and assign them to the df_ratio DataFrame using .loc
df_ratio.loc[:, '比例'] = grouped_data['計數'] / total_counts

# Print the updated df_ratio DataFrame
print(df_ratio)

pivot_df = df_ratio.pivot(index='等級別', columns='日間∕進修別', values='比例')

# Plot a stacked bar chart
ax = pivot_df.plot(kind='bar', stacked=True, figsize=(10, 7))

# Display settings and labels
plt.title('各等級別學制類別比例')
plt.xlabel('等級別')
plt.ylabel('比例')
plt.legend(title='日間∕進修別')
plt.tight_layout()

# Display the plot
plt.show()
```

### 進修別男女比例

```python

df['男生計'] = df['男生計'].astype(str)
df['女生計'] = df['女生計'].astype(str)
df['男生計'] = df['男生計'].str.replace(',-', '').astype(int)
df['女生計'] = df['女生計'].str.replace(',-', '').astype(int)
gender_distribution = df.groupby('日間∕進修別')[['男生計','女生計']].sum().reset_index()

gender_distribution
gender_distribution['Total'] = gender_distribution['男生計'] + gender_distribution['女生計']

gender_ratio = pd.DataFrame(columns = ['日間/進修別','ratio_girl','ratio_boy'])

# gender_ratio['Total'] = gender_distribution['Total']
# gender_ratio['女生計'] = gender_distribution['女生計']
# gender_ratio['男生計'] = gender_distribution['男生計']
gender_ratio['日間/進修別'] = gender_distribution['日間∕進修別']
gender_ratio['ratio_girl'] = gender_distribution['女生計']/gender_distribution['Total']
gender_ratio['ratio_boy'] = gender_distribution['男生計']/gender_distribution['Total']
gender_ratio

gender_ratio.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title('進修別男女比例')
plt.xlabel('日間∕進修別')
plt.ylabel('男女比例')
plt.legend(title='日間∕進修別')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### 所有系學類男女生總數走勢


```python
male_columns = ['一年級男生', '二年級男生', '三年級男生', '四年級男生', '五年級男生', '六年級男生', '七年級男生']

female_columns = ['一年級女生', '二年級女生', '三年級女生', '四年級女生', '五年級女生', '六年級女生', '七年級女生']

# Apply conversion and replacement to each column
for column in male_columns + female_columns:
    df[column] = df[column].astype(str)
    df[column] = df[column].str.replace(',-', '').astype(int)

df_boys = df[male_columns].sum()
df_boys

df_girls = df[female_columns].sum()
df_girls


grade_levels = ['一年級', '二年級', '三年級', '四年級', '五年級', '六年級', '七年級']

# Create a line chart for male and female students
plt.figure(figsize=(10, 6))
plt.plot(grade_levels, df_boys, label='Male')
plt.plot(grade_levels, df_girls, label='Female')
plt.title('Male and Female Student Count by Grade')
plt.xlabel('Grade')
plt.ylabel('Student Count')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

###  各等級別研修生比例

```python
columns = ['延修生女生','延修生男生']

# Apply conversion and replacement to each column
for column in columns:
    df[column] = df[column].astype(str)
    df[column] = df[column].str.replace(',-', '').astype(int)


df_new = df.groupby('等級別')[['延修生女生','延修生男生']].sum().reset_index()
df_new['延修生人數總和'] = df_new['延修生女生'] + df_new['延修生男生']


df_new


plt.figure(figsize=(8, 8))
plt.pie(df_new['延修生人數總和'], labels=df_new['等級別'], autopct='%1.1f%%', startangle=0)
plt.title('各等級別研修生比例')
plt.show()
```

### 各等級別研修生男女佔總男女人數比例

```python
columns = ['男生計','女生計']

# Apply conversion and replacement to each column
for column in columns:
    df[column] = df[column].astype(str)
    df[column] = df[column].str.replace(',-', '').astype(int)


#add total boys and girls count in df_new
df_total = df.groupby('等級別')[['男生計','女生計']].sum().reset_index()
df_new['boys_total'] = df_total['男生計']
df_new['girls_total'] = df_total['女生計']


# calculate ratio 
df_new['boys_ratio'] = df_new['延修生男生']/df_new['boys_total']
df_new['girls_ratio'] = df_new['延修生女生']/df_new['girls_total']

# df_new

plt.figure(figsize=(10, 6))
plt.bar(df_new['等級別'], df_new['boys_ratio'], label='Boys Ratio')
plt.bar(df_new['等級別'], df_new['girls_ratio'], label='Girls Ratio', bottom=df_new['boys_ratio'])
plt.title('Ratio of Boys and Girls by Level')
plt.xlabel('Level')
plt.ylabel('Ratio')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
